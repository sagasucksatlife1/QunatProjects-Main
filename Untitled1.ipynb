```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifie
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
```

```python

tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'JPM', 'V', 'JNJ', 'WMT', 'PG']
market_indices = ['^GSPC', '^DJI', '^IXIC', '^VIX']



all_tickers = tickers + market_indices
start_date = '2020-01-01'
end_date = '2024-12-31'


print("Downloading extended dataset with noise variables...")
data = yf.download(all_tickers, start=start_date, end=end_date)
data.dropna(inplace=True)
print(data.columns)
data.describe()



```

```python
# Calculate returns for all tickers
returns = data['Close'].pct_change()

# Calculate volatility for each ticker
volatility = returns.rolling(window=20).std()

# Flatten the data: combine Close prices and Volatility into a single DataFrame
data_flat = pd.DataFrame()

# Add Close prices
for ticker in all_tickers:
    if ticker in data['Close'].columns:
        data_flat[f'{ticker}_Close'] = data['Close'][ticker]

# Add Volatility
for ticker in all_tickers:
    if ticker in volatility.columns:
        data_flat[f'{ticker}_Volatility'] = volatility[ticker]

# Drop NaN
data_flat = data_flat.dropna()

# Select 5 main tickers (10 columns: Close + Volatility for each)
sample_tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA']
sample_features = [f'{ticker}_Close' for ticker in sample_tickers] + [f'{ticker}_Volatility' for ticker in sample_tickers]

# Pairplot with only 10 columns
sns.pairplot(data_flat[sample_features])
plt.show()
```
